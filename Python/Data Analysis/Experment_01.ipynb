<h3>The matplotlib Library</h3>
<li>It is a Python library specialized in the development of two-dimensional charts (including 3D charts).
<li> Designed to replicate/reproduce an environment very similar to MATLAB (both graphical and syntactic form).
<li>Also exploits optimization techniques that have over the year led to deducibility and simplicity in it's use ~ making it an excellent choice for data visualization (1st timers, without any experience of MATLAB)
    <br><br>
Features that makes matplotlib library stand out:<br>
    1. Extreme simplicity in its use<br>
    2. Gradual development and interactive data visualization<br>
    3. Expressions and text in LaTeX<br>
    4. Greater control over graphic elements<br>
    5. Export to many formats, such as PNG, PDF, SVG, and EPS<br>
   
<h3>The matplotlib Architecture</h3>
<li> Key task - provide a set of fn. & tools that allow representation and manipulation of a figure (main object), along with all internal objects (composition)
<li> Also provide tools for event handeling & the ability to animate graphics</li>
<li> Capable of producing interactive charts - triggered by just one keyboard/mouse click</li>
<li> Architecture of matplotlib - 3 layers (logically) 3 different levels</li>
<li> Communication is unidirectional - each layer can communication with the underlying layer, vice-versa not possible</li>
<li>The three layers are: Scripting --> Artist --> Backend
<br><br>
<b>1.<u> Backend Layer:</u></b><br>
<li>Works at the Lowest level</li>
<li> Contains the matplotlib APIs, a set of classes that play the role of implementation of the graphic elements at a low level</li>
- <i>FigureCanvas</i>: object that emboids the concept of drawing area<br>
- <i>Renderer</i>: object that draws on <i>FigureCanvas</i><br>
- <i>Event</i>: object that handels user input (keyboard and mouse clicks<br>
<br>
<b>2.<u>Artist Layer:</u></b><br>
<li>Intermediate layer, all elements that make up a chart [title, axis labels, markers, etc.] are instance of this layer</li>
<li>Each plays it's role within a hierarchical structure</li>
-<i>Primitive artists:</i> individual objects ~ constitute basic elements to form a graphical representation in a plot[2d line, geometric figures, pieces of text]<br>
-<i>Composite artists:</i> graphic elements in chart ~ composed of several base elements (primitive) [ axis, ticks, axes and figures]<br>
<i> 1. <u>Figure:</u> the object with highest level in hierarchy ~ corresponding to entire graphical representation, contain many axis<br>
    2.<u>Axes: </u> plot/chart, each axis object belongs to only one figure. title, x label, y label are part of this<br>
    3.<u>Axis objects:</u> accounts numerical value to be represented on Axes. Defines the limits, manages the ticks & tick labels. Position of tick managed by <b> locator</b> fromatting of tick by <b> formatter</b></i>
    <br><br>
<b> 3.<u> Scripting Layer:</u></b><br>
<li> For the purpose of calculation, analysis and visualization of data, this layer is the best.</li>
<li> Consists of <i>pyplot</i></li><br>

<h3> Working with Multiple Figures and Axes </h3>
<li> matplotlib allows you to manage multiple figures simulatneously ~ within each feature,a allows you to view different plots as subplots</li>
<li>The <i>subplot()</i> fn. ~ subdividing the figure in different drawing areas + focus the commands on a specific subplot</li>
<li>Argument passed to <i>subplot()</i> fn., sets the mode of subdivision and determines which is the currrent subplot, then the current subplot will be the only figure that will be affected by the commands</li>
<li> Argument of <i>subplot()</i> composed of 3 integers:<br> 1st - how many parts the figure is split into vertically<br>
    2nd - how many parts the figure is divided into horizontally<br>
    3rd - selects the current subplot on which the commands are directed<br>

```python
#dividing the canvas into two horizontal subplots
import numpy as np
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)

p.subplot(211)
p.plot(t,y1,'b-.')

p.subplot(212)
p.plot(t,y2,'r--')
```

```python
#dividing the canvas into two vertical subplots
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)

p.subplot(121)
p.plot(t,y1,'b-.')

p.subplot(122)
p.plot(t,y2,'r--')
```

<h3> Adding Elements to the Chart </h3>

<h4>1. Adding Text </h4>
<li> <i>title()</i> fn. ~ already discussed</li>
<li> <i>xlabel & ylabel</i> fn. taken as string arguments

```python
p.axis([0,5,0,20])
p.title('My first plot',fontsize=20,fontname='Times New Roman')
#changing the font style
p.xlabel('Counting',color='gray')
p.ylabel('Square values',color='gray')
#p.plot([1,2,3,4],[1,4,9,16],'ro')
p.text(1,1.5,'First') #adding labels to each pointers
p.text(2,4.5,'Second')
p.text(3,9.5,'Third')
p.text(4,16.5,'Fourth')
p.plot([1,2,3,4],[1,4,9,16],'ro')
```

<li> matplotlib offers the possibility to integrate LaTeX expressions ~ allowing you to insert mathematical equations with the chart</li>
<li> LaTeX eqn. to text enclosed between two $ charachters ~ interpreter will recognize it as LaTeX and convert them to corresponding graphic [mathematical eq., a formula, mathemtical characters, greek letters]</li>
<li>precede the LaTeX expression with 'r' = raw text ~ avoid unintended escape sequences</li>

```python
p.axis([0,5,0,20])
p.title('My first plot',fontsize=20,fontname='Times New Roman')
#changing the font style
p.xlabel('Counting',color='gray')
p.ylabel('Square values',color='gray')
p.text(1,1.5,'First') #adding labels to each pointers
p.text(2,4.5,'Second')
p.text(3,9.5,'Third')
p.text(4,16.5,'Fourth')
p.text(1.1,12,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
#enclosing the equation in a coloured box
p.plot([1,2,3,4],[1,4,9,16],'ro')
```

<h3>Handling Date Values</h3>
<li>One of the most common problems encountered while data analysis ~ handling data of date-time type</li>
<li>Displaying the data along the axis can be problematic, especially when managing ticks, makes it difficult to read</li>
<li>Hence to manage dates, define a time scale with appropriate objects, for which you need to import <i>matplotlib.dates</i> module</li>
<li> Then you define the scales of days with <i>DayLocator()</i> and for months with <i>MonthLocator</i> fns.</li>
<li>Formatting is improtant ~ avoid overlap, limit the tick labels to the essential [yy-mm-dd], can be done with <i>DateFormatter()</i> fn.</li>
<li>After defining the scles you can set two different kinds of ticks on the x-axis using <i> set_major_locator() & set_minor_locator </i> fn. </li>
<li> To set teh text format of the tick lables referred to the month use <i>set_major_formatter()</i></li>

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')
events = [datetime.date(2015,1,23),datetime.date(2015,1,28),datetime.date(2015,2,3),datetime.date(2015,2,21),datetime.date(2015,3,15),datetime.date(2015,3,24),datetime.date(2015,4,8),datetime.date(2015,4,24)]
readings = [12,22,25,20,18,15,17,14]
fig, ax = plt.subplots()
plt.plot(events,readings)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
```

<h3> Bar Charts </h3>
<li>Similar to histogram, but e x-axis is not used to reference numerical values but categories</li>
<li> You can specify the category throught tick labels, defined by a list of strings passed to the <i>xticks()</i> fn.</li>
<li> You can add standard deviation using <i> yerr</i> with a list containing the same</li>
<li> <i>error_kw</i> error bars,<i>eColor</i> specifies the color of the error bar, <i>capsize</i> defines the width of the transverse lines that mar the end of the error bars, <i>alpha </i> indicates the degree of transparency of the colored bar [value ranging from 0-1](0 - completely transparent, 1 - fully)</li>

```python
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('A Bar Chart')
plt.bar(index,values1,yerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='First')
plt.xticks(index+0.4,['A','B','C','D','E'])
plt.legend(loc=2)
```

<h4> Horizontal Bar Charts </h4>
The only change that you have to take into account is that the roles of the axes are 
reversed.

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('A Horizontal Bar Chart')
plt.barh(index,values1,xerr=std1,error_kw={'ecolor':'0.1', 'capsize':6},alpha=0.7,label='First')
plt.yticks(index+0.4,['A','B','C','D','E'])
plt.legend(loc=5)
```

<h4> Multiserial Bar Charts </h4>
<li> Bar charts are also generally used to simultaneously display larger series of values</li>
<li> You have more bars that must share the same category</li>
<li> Advisable to add space, gap to separate a category with respect to the next</li>

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,5,0,8])
plt.title('A Multiseries Bar Chart',fontsize=20)
plt.bar(index,values1,bw,color='b')
plt.bar(index+bw,values2,bw,color='g')
plt.bar(index+2*bw,values3,bw,color='r')
plt.xticks(index+1.5*bw,['A','B','C','D','E'])
```

<h4> Multiseries Stacked Bar Charts </h4>
<li> Useful when you want to show the total value obtained by the sum of all the bars </li>
<li>To transform a simple multiseries bar chart in a stacked one, you add the bottom kwarg to each bar() function</li>
<li>Each series must be assigned to the corresponding bottom kwarg</li>
<li>At the end you will obtain the stacked bar chart</li>

```python
import matplotlib.pyplot as plt
import numpy as np
series1 = np.array([3,4,5,3])
series2 = np.array([1,2,2,5])
series3 = np.array([2,3,3,4])
index = np.arange(4)
plt.axis([-0.5,3.5,0,15])
plt.title('A Multiseries Stacked Bar Chart')
plt.bar(index,series1,color='r')
plt.bar(index,series2,color='b',bottom=series1)
plt.bar(index,series3,color='g',bottom=(series2+series1))
plt.xticks(index+0.4,['Jan18','Feb18','Mar18','Apr18'])
```

```python
plt.axis([0,15,-0.5,3.5])
plt.title('A Multiseries Horizontal Stacked Bar Chart')
plt.barh(index,series1,color='r')
plt.barh(index,series2,color='g',left=series1)
plt.barh(index,series3,color='b',left=(series1+series2))
plt.yticks(index+0.4,['Jan18','Feb18','Mar18','Apr18'])
```

```python
# Another way of distinguishing the series in bar chart
plt.title('A Multiseries Horizontal Stacked Bar Chart')
plt.barh(index,series1,color='w',hatch='xx')
plt.barh(index,series2,color='w',hatch='///', left=series1)
plt.barh(index,series3,color='w',hatch='\\\\\\',left=(series1+series2))
plt.yticks(index+0.4,['Jan18','Feb18','Mar18','Apr18'])
```

<h4> 3D Surfaces </h4>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1, cmap=plt.cm.hot)
ax.view_init(elev=30,azim=125)
```

<h3> Multi-Panel Plots </h3>
Displaying Subplots Within Other Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
inner_ax = fig.add_axes([0.6,0.6,0.25,0.25])
x1 = np.arange(10)
y1 = np.array([1,2,7,1,5,2,4,2,3,1])
x2 = np.arange(10)
y2 = np.array([1,3,4,5,4,5,2,6,4,3])
ax.plot(x1,y1)
inner_ax.plot(x2,y2)
```
