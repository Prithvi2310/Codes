```python
t = np.arange(0.,1.,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
plt.subplot(121)
plt.plot(t,y1,'r')
plt.subplot(122)
plt.plot(t,y2,'g')
plt.title("Math Functions",fontsize=30,font="cursive",color="b")
```
