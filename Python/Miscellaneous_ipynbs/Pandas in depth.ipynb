# 1. Using group by function

```python
import pandas as p
data = {'Name':['Jai', 'Anuj', 'Jai', 'Princi','Gaurav', 'Anuj', 'Princi', 'Abhi'],'Age':[27, 24, 22, 32,33, 36, 27, 32],'College':['SRM','SRM','VIT','VIT','IIT','IIT','NIT','NIT'],'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj','Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],'Qualification':['Msc', 'MA', 'MCA', 'Phd','B.Tech', 'B.com', 'Msc', 'MA']}
df = p.DataFrame(data)
print(df)
```

```python
df.groupby('College')
print(df.groupby('College').groups)
```

```python
a = df.groupby('College')
a.first()
```

```python
df.groupby(['College', 'Qualification']) 
print(df.groupby(['College', 'Qualification']).groups)
```

# 2. Pivoting
To rearrange the data ie converting rows into columns and columns into rows.
Two operations:
1) Stacking
2) Unstacking


```python
# import pandas module
import pandas as pd
# making dataframe
df1 = pd.read_csv("D:/SRMIST/Foundation of Data Analysis/nba.csv")
# it will print the first 5-rows
df1.head()
```

```python
# reshape the dataframe using stack() method  [rows into columns]
df1_stacked = df1.stack()   
print(df1_stacked.head(26))
```

```python
#unstack()(columns ito rows)
print(df1_stacked.unstack())
```

# 4. Data Transformation

```python
#Use DataFrame.transform() function to add 10 to each element in the dataframe.
# importing pandas as p
df2 = p.DataFrame({"A":[12, 4, 5, None, 1], "B":[7, 2, 54, 3, None], "C":[20, 16, 11, 3, 8], "D":[14, 3, None, 2, 6]}) 
# Create the index
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']
# Set the index
df2.index = index_
# Print the DataFrame
df2
```

```python
# add 10 to each element of the dataframe
result = df2.transform(func = lambda x : x + 10)
# Print the result
result
```

# 6. Mapping

```python
frame1 = p.DataFrame({ 'item':['ball','mug','pen','pencil','ashtray'],'color':['white','rosso','verde','black','yellow'],'price':[5.56,4.20,1.30,0.56,2.75]})
frame1
```

```python
# Replacing values through Mapping
newcolors = {'rosso': 'red','verde': 'green'}
frame1.replace(newcolors)
```

```python
ser = p.Series([1,3,n.nan,4,6,n.nan,3])
print(ser)
print(" ")
print(ser.replace(n.nan,0))
```

```python
#adding values using mapping
frame2 = pd.DataFrame({'item':['ball','mug','pen','pencil','ashtray'],'color':['white','red','green','black','yellow']})
frame2
```

```python
price = {'ball': 5.56,'mug': 4.20,'bottle': 1.30,'scissors': 3.41,'pen': 1.30,'pencil': 0.56,'ashtray': 2.75}
frame2['price'] = frame2['item'].map(price)
frame2
```

# 8. Discretization/Binning
The process of converting a continuous variable to a discrete variable by creating a set of contiguous intervals that span the range of the variable's value.<br>
Discretization is also called binning, where bin is an alternative name for interval.<br>

<h4> Why use Discretization</h4>
<li>To handle outliners</li>
<li>To improve the speed of producing values</li>


```python
# equal frequency
def equifreq(arr1, m):   
    a = len(arr1)
    n = int(a / m)
    for i in range(0, m):
        arr = []
        for j in range(i * n, (i + 1) * n):
            if j >= a:
                break
            arr = arr + [arr1[j]]
        print(arr)
 
# equal width
def equiwidth(arr1, m):
    a = len(arr1)
    w = int((max(arr1) - min(arr1)) / m)
    min1 = min(arr1)
    arr = []
    for i in range(0, m + 1):
        arr = arr + [min1 + w * i]
    arri=[]
     
    for i in range(0, m):
        temp = []
        for j in arr1:
            if j >= arr[i] and j <= arr[i+1]:
                temp += [j]
        arri += [temp]
    print(arri)
 
# data to be binned
data4 = [5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215] 
# no of bins
m = 3 
print("equal frequency binning")
equifreq(data4, m)

print("\n\nequal width binning")
equiwidth(data4, 3)
```
